 **CONDITIONAL STATEMENTS**

```python
#18+ drive else cannot drive

age = int(input("Enter your age: "))
print("Your age is " , age)

if(age>=18):
  print("You can drive a vehicle")

else:
  print("You cannot drive a vehicle")
```

```python
#Write a program to check whether a number is positive, negative, or zero

number = int(input("Enter a number "))

if(number<0):
  print("Negative number")

elif(number>0):
  print("Positive number")
else:
  print("Zero")
```

```python
#Take a student's marks as input and print:
#A if marks ≥ 90
#B if marks ≥ 75
#C if marks ≥ 50
#Fail otherwise

marks = int(input("Enter marks you got"))

if (marks>=90):
  print("A")
elif (marks>=75):
  print("B")
elif (marks>=50):
  print("C")
else:
  print("Fail")

```

```python
#Check whether a number is even or odd.

number = int(input("Enter number "))

if (number%2==0):
  print("Even")
else:
  print("Odd")
```

```python
#Take age as input and check:
#below 18 → “Minor”
#18 to 60 → “Adult”
#above 60 → “Senior Citizen”

age = int(input("Enter age "))

if (age<18):
  print("Minor")
elif (age<=60):
  print("Adult")
else:
  print("Senior citizen")
```

```python
#Write a program to find the largest among 3 numbers.

a = int(input("Enter 1st number"))
b = int(input("Enter 2nd number"))
c = int(input("Enter 3rd number"))

if (a>b and a>c):
  print("A is the largest number")
elif (b>a and b>c):
  print("B is the largest number")
else:
  print("C is the largest number")
```

```python
#Check whether a year is a leap year or not.
#The year is multiple of 400.
#The year is a multiple of 4 and not a multiple of 100.

year = int(input("Enter year: "))

if (year%400 == 0 or year % 4 ==0 and year % 100 != 0):
  print("Leap Year")

else:
  print("Not a leap year")
```

```python
#Take username and password input and check if login is successful.
#username = admin
#password = 1234

username = input("Enter username: ")
password = input("Enter password: ")

if (username == "admin" and password == "1234"):
  print("Login Successful")
else:
  print("Login unsuccessful")
```

```python
#Check whether a character is:
#uppercase
#lowercase
#digit
#special character

ch = input("Enter character: ")

if (ch.isupper()):
  print("Upper case")
elif (ch.islower()):
  print("Lower case")
elif (ch.isdigit()):
  print("digit")
else:
  print("Special character")

```

```python
#without using function

ch = input("Enter character: ")

if (len(ch) ==1):
  if (ch >= 'a' and ch <= 'z'):
    print("Lowercase")

  elif (ch >= 'A' and ch <= 'Z'):
    print("Upper case")

  elif (ch >= '0' and ch <= '9'):
    print("Number")

  else:
    print("Special character")

else:
  print("Enter 1 character only not more than that")
```

```python
#Take a temperature input and print:
#“Cold” if below 15
#“Normal” if 15–30
#“Hot” if above 30

temp = int(input("Enter temperature in C  "))

if (temp<15):
  print("Cold")

elif (temp<30):
  print("Normal")

else:
  print("Hot")
```
